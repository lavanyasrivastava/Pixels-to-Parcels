```python
# Part 5 — Linear Regression
# Dataset: Advertising Dataset (TV, Radio, Newspaper → Sales)
# ============================================================

# ── IMPORTS ──────────────────────────────────────────────────
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


# ── LOAD DATASET ─────────────────────────────────────────────
file_path = ('/content/sample_data/advertising.csv')
df = pd.read_csv(file_path)

print("=" * 55)
print("Dataset Preview")
print("=" * 55)
print(df.head())
print(f"\nShape: {df.shape}")
print(df.describe().round(2))


# ============================================================
# Q5.1 — SCRATCH IMPLEMENTATION: Simple Linear Regression

# ============================================================

# 1. Manually split: first 160 rows = train, remaining = test
X_scratch = df['TV'].values       # feature: TV advertising budget
y_scratch  = df['Sales'].values   # target:  Sales

X_train_s = X_scratch[:160]       # first 160 rows → train
y_train_s  = y_scratch[:160]

X_test_s  = X_scratch[160:]       # remaining rows → test
y_test_s   = y_scratch[160:]

print("\n" + "=" * 55)
print("Q5.1 — Manual Train/Test Split")
print("=" * 55)
print(f"Train size : {len(X_train_s)}")
print(f"Test size  : {len(X_test_s)}")

# 2. Compute slope (m) and intercept (b) using closed-form formula
#    m = Σ((xi - x̄)(yi - ȳ)) / Σ((xi - x̄)²)
#    b = ȳ - m × x̄

x_mean = np.mean(X_train_s)       # mean of X (TV)
y_mean = np.mean(y_train_s)       # mean of y (Sales)

numerator   = np.sum((X_train_s - x_mean) * (y_train_s - y_mean))  # Σ(xi-x̄)(yi-ȳ)
denominator = np.sum((X_train_s - x_mean) ** 2)                     # Σ(xi-x̄)²

m = numerator / denominator        # slope
b = y_mean - m * x_mean            # intercept

# 3. Print computed m and b
print("\n" + "=" * 55)
print("Q5.1 — Scratch Computed Parameters")
print("=" * 55)
print(f"Slope (m)     : {m:.6f}")
print(f"Intercept (b) : {b:.6f}")
print(f"\nEquation: Sales = {m:.4f} × TV + {b:.4f}")

# 4. Generate predictions on test set
y_pred_scratch = m * X_test_s + b   # ŷ = m*x + b

print("\nSample Predictions vs Actual (first 5 test rows):")
for i in range(5):
    print(f"  TV={X_test_s[i]:.1f} → Predicted={y_pred_scratch[i]:.2f}, Actual={y_test_s[i]:.2f}")

# 5. Compute MSE manually (no sklearn)
mse_scratch = np.mean((y_test_s - y_pred_scratch) ** 2)   # MSE = (1/n) Σ(yi - ŷi)²

print(f"\nManual MSE (scratch) on Test Set : {mse_scratch:.4f}")

# Plot scratch regression line on training data
plt.figure(figsize=(8, 5))
plt.scatter(X_train_s, y_train_s, color='steelblue', alpha=0.6,
            label='Train Data', edgecolors='white', linewidths=0.3, s=50)
plt.scatter(X_test_s,  y_test_s,  color='orange', alpha=0.7,
            label='Test Data',  edgecolors='white', linewidths=0.3, s=50)
x_line = np.linspace(X_scratch.min(), X_scratch.max(), 200)
plt.plot(x_line, m * x_line + b, color='red', linewidth=2,
         label=f'Scratch Line: y={m:.4f}x+{b:.4f}')
plt.title('Q5.1 — Scratch Linear Regression (TV → Sales)')
plt.xlabel('TV Advertising Budget')
plt.ylabel('Sales')
plt.legend()
plt.grid(linestyle='--', alpha=0.4)
plt.tight_layout()
plt.savefig('scratch_regression.png', dpi=150)
plt.show()
```

```python
# 5.2)
# Q5.2 — SKLEARN IMPLEMENTATION
# ============================================================

# 1. Split data 80-20 using train_test_split (random_state=42)
X_sk = df[['TV']]             # 2D array needed for sklearn
y_sk = df['Sales']

X_train_sk, X_test_sk, y_train_sk, y_test_sk = train_test_split(
    X_sk, y_sk, test_size=0.2, random_state=42
)

print("\n" + "=" * 55)
print("Q5.2 — Sklearn Train/Test Split")
print("=" * 55)
print(f"Train size : {len(X_train_sk)}")
print(f"Test size  : {len(X_test_sk)}")

# 2. Train LinearRegression model from sklearn on TV → Sales
sk_model = LinearRegression()
sk_model.fit(X_train_sk, y_train_sk)

# 3. Print coefficient (slope) and intercept
sk_m = sk_model.coef_[0]       # slope
sk_b = sk_model.intercept_     # intercept

print("\n" + "=" * 55)
print("Q5.2 — Sklearn Model Parameters")
print("=" * 55)
print(f"Sklearn Coefficient (m) : {sk_m:.6f}")
print(f"Sklearn Intercept   (b) : {sk_b:.6f}")

# Compare with scratch values
print("\n── Comparison: Scratch vs Sklearn ──────────────────")
print(f"{'Parameter':<20} {'Scratch':>12} {'Sklearn':>12} {'Difference':>12}")
print("-" * 56)
print(f"{'Slope (m)':<20} {m:>12.6f} {sk_m:>12.6f} {abs(m-sk_m):>12.6f}")
print(f"{'Intercept (b)':<20} {b:>12.6f} {sk_b:>12.6f} {abs(b-sk_b):>12.6f}")

# 4. Why might there be a small difference?
print("""
── Why the small difference? ───────────────────────────
The scratch implementation splits the first 160 rows sequentially
(rows 0–159 = train), while sklearn's train_test_split(random_state=42)
randomly shuffles the data before splitting 80/20. Because different rows
end up in the training set, the means (x̄, ȳ) differ slightly, which
propagates to slightly different m and b values. Both are mathematically
correct — they just trained on different subsets of the data.
""")

```

```python
# Q5.3 — EVALUATION METRICS
# ============================================================

# Generate predictions from sklearn model for both sets
y_train_pred_sk = sk_model.predict(X_train_sk)
y_test_pred_sk  = sk_model.predict(X_test_sk)

# ── Helper function to compute all 4 metrics ─────────────────
def compute_metrics(y_true, y_pred, set_name):
    mae  = mean_absolute_error(y_true, y_pred)
    mse  = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)                          # RMSE = √MSE
    r2   = r2_score(y_true, y_pred)
    print(f"\n── {set_name} ──────────────────────────────────")
    print(f"  MAE  (Mean Absolute Error)       : {mae:.4f}")
    print(f"  MSE  (Mean Squared Error)        : {mse:.4f}")
    print(f"  RMSE (Root Mean Squared Error)   : {rmse:.4f}")
    print(f"  R²   Score                       : {r2:.4f}")
    return {'Set': set_name, 'MAE': round(mae,4), 'MSE': round(mse,4),
            'RMSE': round(rmse,4), 'R2': round(r2,4)}

print("\n" + "=" * 55)
print("Q5.3 — Evaluation Metrics")
print("=" * 55)

train_metrics = compute_metrics(y_train_sk, y_train_pred_sk, "TRAIN SET")
test_metrics  = compute_metrics(y_test_sk,  y_test_pred_sk,  "TEST SET")

# Print as a clean comparison DataFrame
metrics_df = pd.DataFrame([train_metrics, test_metrics]).set_index('Set')
print("\n── Metrics Summary Table ───────────────────────────")
print(metrics_df.to_string())

# Plot: Actual vs Predicted on test set
plt.figure(figsize=(8, 5))
plt.scatter(y_test_sk, y_test_pred_sk, color='steelblue', alpha=0.7,
            edgecolors='white', linewidths=0.3, s=60, label='Test Predictions')
min_val = min(y_test_sk.min(), y_test_pred_sk.min())
max_val = max(y_test_sk.max(), y_test_pred_sk.max())
plt.plot([min_val, max_val], [min_val, max_val],
         'r--', linewidth=2, label='Perfect Prediction Line')
plt.title('Q5.3 — Actual vs Predicted Sales (Test Set)')
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.legend()
plt.grid(linestyle='--', alpha=0.4)
plt.tight_layout()
plt.savefig('actual_vs_predicted.png', dpi=150)
plt.show()

# Residuals plot
residuals = y_test_sk.values - y_test_pred_sk
plt.figure(figsize=(8, 4))
plt.scatter(y_test_pred_sk, residuals, color='purple', alpha=0.6,
            edgecolors='white', linewidths=0.3, s=55)
plt.axhline(0, color='red', linestyle='--', linewidth=1.5)
plt.title('Residuals Plot (Test Set)')
plt.xlabel('Predicted Sales')
plt.ylabel('Residuals (Actual − Predicted)')
plt.grid(linestyle='--', alpha=0.4)
plt.tight_layout()
plt.savefig('residuals_plot.png', dpi=150)
plt.show()


```

```python
# Q. 6)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')


# ============================================================
# Q6.1 — DATA EXPLORATION
# ============================================================

# 1. Load the dataset
df = pd.read_csv('Mall_Customers.csv')

# Display .head()
print("=" * 60)
print("HEAD — First 5 rows")
print("=" * 60)
print(df.head())

# Display .info()
print("\n" + "=" * 60)
print("INFO — Column types & null counts")
print("=" * 60)
df.info()

# Display .describe()
print("\n" + "=" * 60)
print("DESCRIBE — Summary statistics")
print("=" * 60)
print(df.describe().round(2))


# ── Rename columns for easier access ─────────────────────────
df.rename(columns={
    'Annual Income (k$)'     : 'AnnualIncome',
    'Spending Score (1-100)' : 'SpendingScore'
}, inplace=True)


# 2. Scatter plot — Annual Income vs Spending Score (RAW)
plt.figure(figsize=(7, 5))
plt.scatter(df['AnnualIncome'], df['SpendingScore'],
            color='steelblue', alpha=0.7, edgecolors='white', linewidths=0.4, s=60)
plt.title('Annual Income vs Spending Score (Raw Data)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1–100)')
plt.grid(linestyle='--', alpha=0.4)
plt.tight_layout()
plt.savefig('scatter_raw.png', dpi=150)
plt.show()
# Visual observation: ~5 natural clusters visible —
# four corner groups (high/low income x high/low spending)
# + one central average group.  Guess → k = 5


# ============================================================
# Q6.2 — APPLY K-MEANS
# ============================================================

# Feature matrix (only the two columns used for clustering)
X = df[['AnnualIncome', 'SpendingScore']].values


# ── Step 1: Find optimal k using the Elbow Method ────────────
inertias = []
K_range  = range(1, 11)

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X)
    inertias.append(km.inertia_)   # inertia = WCSS

plt.figure(figsize=(7, 4))
plt.plot(list(K_range), inertias, 'o-', color='tomato', linewidth=2, markersize=7)
plt.axvline(5, color='green', linestyle='--', linewidth=1.5, label='Optimal k = 5')
plt.title('Elbow Method — Finding Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (WCSS)')
plt.legend()
plt.grid(linestyle='--', alpha=0.4)
plt.tight_layout()
plt.savefig('elbow_curve.png', dpi=150)
plt.show()
# Elbow clearly bends at k=5 → optimal k = 5


# ── Step 2: Fit K-Means with k = 5 ───────────────────────────
optimal_k = 5

kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X)   # adds Cluster column

print("\n" + "=" * 60)
print("Dataframe with Cluster column (.head())")
print("=" * 60)
print(df.head(10))


# ── Step 3: Print cluster centroids ──────────────────────────
centroids = pd.DataFrame(
    kmeans.cluster_centers_,
    columns=['Annual Income (k$)', 'Spending Score'],
    index=[f'Cluster {i}' for i in range(optimal_k)]
)

print("\n" + "=" * 60)
print("CLUSTER CENTROIDS")
print("=" * 60)
print(centroids.round(2).to_string())


# ── Step 4: Per-cluster summary ───────────────────────────────
summary = df.groupby('Cluster').agg(
    Count               = ('Cluster',       'count'),
    Mean_Annual_Income  = ('AnnualIncome',   'mean'),
    Mean_Spending_Score = ('SpendingScore',  'mean')
).round(2)

print("\n" + "=" * 60)
print("PER-CLUSTER SUMMARY")
print("=" * 60)
print(summary.to_string())

# Loop version — prints each cluster individually
print("\n── Individual Cluster Stats ──────────────────────────")
for cluster_id, row in summary.iterrows():
    print(f"\nCluster {cluster_id}:")
    print(f"  Count of customers   : {int(row['Count'])}")
    print(f"  Mean Annual Income   : ${row['Mean_Annual_Income']:.2f}k")
    print(f"  Mean Spending Score  : {row['Mean_Spending_Score']:.2f}")


# ── Step 5: Final clustered scatter plot ─────────────────────
COLORS = ['#E63946', '#2EC4B6', '#F4A261', '#8338EC', '#06D6A0']

plt.figure(figsize=(8, 6))

for c in range(optimal_k):
    mask = df['Cluster'] == c
    plt.scatter(df.loc[mask, 'AnnualIncome'],
                df.loc[mask, 'SpendingScore'],
                color=COLORS[c], alpha=0.85, s=60,
                edgecolors='white', linewidths=0.3,
                label=f'Cluster {c}')

# Plot centroids as stars
cx = kmeans.cluster_centers_[:, 0]
cy = kmeans.cluster_centers_[:, 1]
plt.scatter(cx, cy, marker='*', s=300, color='white',
            edgecolors='black', linewidths=0.8,
            zorder=10, label='Centroids')

plt.title(f'K-Means Clustering  (k = {optimal_k})')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1–100)')
plt.legend(loc='upper left', fontsize=9)
plt.grid(linestyle='--', alpha=0.4)
plt.tight_layout()
plt.savefig('kmeans_clustered.png', dpi=150)
plt.show()


```

Visual groupings (raw scatter): The plot clearly reveals 5 natural clusters — four corner blobs (high/low income × high/low spending) and one central mass. The elbow curve confirms k=5 as the optimal knee point.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
from sklearn.metrics import confusion_matrix
import warnings
warnings.filterwarnings('ignore')



# Q7.1 — DATA LOADING & PREPROCESSING

# 1. Load the MNIST dataset (built into Keras — no download needed)
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Print shapes
print("=" * 50)
print("Dataset Shapes")
print("=" * 50)
print(f"X_train : {X_train.shape}")   # (60000, 28, 28)
print(f"y_train : {y_train.shape}")   # (60000,)
print(f"X_test  : {X_test.shape}")    # (10000, 28, 28)
print(f"y_test  : {y_test.shape}")    # (10000,)

# 2. Normalize pixel values from [0, 255] → [0, 1]
X_train = X_train / 255.0
X_test  = X_test  / 255.0

print("\nPixel values normalized to [0, 1]")
print(f"Min: {X_train.min():.1f}  |  Max: {X_train.max():.1f}")

# 3. Display a 5×5 grid of 25 random training images with true labels
fig, axes = plt.subplots(5, 5, figsize=(8, 8))
fig.suptitle('25 Random MNIST Training Images', fontsize=14, fontweight='bold')

random_indices = np.random.choice(len(X_train), 25, replace=False)

for i, ax in enumerate(axes.flat):
    ax.imshow(X_train[random_indices[i]], cmap='gray')   # show image
    ax.set_title(f"Label: {y_train[random_indices[i]]}", fontsize=9)
    ax.axis('off')                                        # turn off axes

plt.tight_layout()
plt.savefig('mnist_sample_grid.png', dpi=150)
plt.show()
```

**Architecture:**
- Input: Flatten (28×28 → 784)
- → Dense(128, relu)
- → Dense(64, relu)
- → Dense(10, softmax)

Compiled with `adam` optimizer and `sparse_categorical_crossentropy` loss.  
Trained for **15 epochs** with `validation_split=0.2`.

```python
# Architecture:
#   Flatten (28×28 → 784)
#   → Dense(128, relu)
#   → Dense(64,  relu)
#   → Dense(10,  softmax)

# Build the model
ffnn_model = keras.Sequential([
    layers.Flatten(input_shape=(28, 28)),          # flatten 28x28 image to 784
    layers.Dense(128, activation='relu'),           # hidden layer 1
    layers.Dense(64,  activation='relu'),           # hidden layer 2
    layers.Dense(10,  activation='softmax')         # output: 10 digit classes
], name="Feedforward_NN")

# 1. Compile the model
ffnn_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Print model summary — note total parameters
print("\n" + "=" * 50)
print("FEEDFORWARD NN — Model Summary")
print("=" * 50)
ffnn_model.summary()

# 2. Count total parameters
ffnn_params = ffnn_model.count_params()
print(f"\nTotal Parameters (FFNN): {ffnn_params:,}")

# 3. Train for 15 epochs with validation_split=0.2
print("\nTraining Feedforward NN...")
ffnn_history = ffnn_model.fit(
    X_train, y_train,
    epochs=15,
    validation_split=0.2,    # 20% of training data used for validation
    verbose=1
)

# 4a. Plot Training & Validation ACCURACY (separate graph)
plt.figure(figsize=(8, 5))
plt.plot(ffnn_history.history['accuracy'],     label='Train Accuracy',      color='steelblue', linewidth=2)
plt.plot(ffnn_history.history['val_accuracy'], label='Validation Accuracy', color='orange',    linewidth=2, linestyle='--')
plt.title('Feedforward NN — Training & Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('ffnn_accuracy.png', dpi=150)
plt.show()

# 4b. Plot Training & Validation LOSS (separate graph)
plt.figure(figsize=(8, 5))
plt.plot(ffnn_history.history['loss'],     label='Train Loss',      color='steelblue', linewidth=2)
plt.plot(ffnn_history.history['val_loss'], label='Validation Loss', color='orange',    linewidth=2, linestyle='--')
plt.title('Feedforward NN — Training & Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('ffnn_loss.png', dpi=150)
plt.show()

```

1. Test accuracy & loss
2. Confusion matrix (sklearn)
3. Heatmap visualization (plt.imshow)
4. Most confused digit pair + 3 misclassified examples

```python
# 1. Print final test accuracy and test loss
print("\n" + "=" * 50)
print("FEEDFORWARD NN — Test Evaluation")
print("=" * 50)
ffnn_test_loss, ffnn_test_acc = ffnn_model.evaluate(X_test, y_test, verbose=0)
print(f"Test Loss     : {ffnn_test_loss:.4f}")
print(f"Test Accuracy : {ffnn_test_acc:.4f}  ({ffnn_test_acc*100:.2f}%)")

# 2. Generate predictions on the test set
y_pred_probs = ffnn_model.predict(X_test)           # shape: (10000, 10) — probabilities
y_pred       = np.argmax(y_pred_probs, axis=1)      # take class with highest probability

# Build confusion matrix using sklearn
cm = confusion_matrix(y_test, y_pred)

print("\nConfusion Matrix (rows=True label, cols=Predicted label):")
print(cm)

# 3. Visualize the confusion matrix as a heatmap using plt.imshow
fig, ax = plt.subplots(figsize=(9, 7))
im = ax.imshow(cm, interpolation='nearest', cmap='Blues')
plt.colorbar(im, ax=ax)

# Add axis labels
ax.set_xticks(np.arange(10))
ax.set_yticks(np.arange(10))
ax.set_xticklabels(range(10))
ax.set_yticklabels(range(10))
ax.set_xlabel('Predicted Label', fontsize=12)
ax.set_ylabel('True Label',      fontsize=12)
ax.set_title('Confusion Matrix — Feedforward NN', fontsize=14, fontweight='bold')

# Annotate each cell with its count
for i in range(10):
    for j in range(10):
        color = 'white' if cm[i, j] > cm.max() / 2 else 'black'
        ax.text(j, i, str(cm[i, j]), ha='center', va='center',
                color=color, fontsize=9, fontweight='bold')

plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=150)
plt.show()

# 4. Find the two most commonly confused digit pairs
cm_no_diag = cm.copy()
np.fill_diagonal(cm_no_diag, 0)           # zero out correct predictions
most_confused_idx = np.unravel_index(np.argmax(cm_no_diag), cm_no_diag.shape)
true_digit  = most_confused_idx[0]
pred_digit  = most_confused_idx[1]
count       = cm_no_diag[true_digit, pred_digit]

print("\n" + "=" * 50)
print("Most Confused Digit Pair")
print("=" * 50)
print(f"True digit {true_digit} was predicted as {pred_digit}  →  {count} times")
print(f"\nWhy? Digits {true_digit} and {pred_digit} share similar curve/stroke patterns")
print("that a flat Dense layer struggles to distinguish spatially.")

# Show 3 example misclassified images
misclassified_mask    = (y_pred != y_test)                     # boolean mask of all errors
misclassified_indices = np.where(misclassified_mask)[0]        # indices of errors

print(f"\nTotal misclassified: {misclassified_mask.sum()} / {len(y_test)}")

fig, axes = plt.subplots(1, 3, figsize=(10, 3))
fig.suptitle('3 Example Misclassified Images (FFNN)', fontsize=13, fontweight='bold')

for i, ax in enumerate(axes):
    idx = misclassified_indices[i]
    ax.imshow(X_test[idx], cmap='gray')
    ax.set_title(f"True: {y_test[idx]}  |  Pred: {y_pred[idx]}", fontsize=10, color='red')
    ax.axis('off')

plt.tight_layout()
plt.savefig('misclassified_examples.png', dpi=150)
plt.show()
```

```python

# Q7.4 — CONVOLUTIONAL NEURAL NETWORK (CNN)

# Reshape X_train and X_test for CNN input: add channel dimension (grayscale = 1)
X_train_cnn = X_train.reshape(-1, 28, 28, 1)   # shape: (60000, 28, 28, 1)
X_test_cnn  = X_test.reshape(-1, 28, 28, 1)    # shape: (10000, 28, 28, 1)

print("\n" + "=" * 50)
print("CNN Input Shapes")
print("=" * 50)
print(f"X_train_cnn : {X_train_cnn.shape}")
print(f"X_test_cnn  : {X_test_cnn.shape}")

# Build the CNN model (exact architecture as specified)
cnn_model = keras.Sequential([
    # Input layer — shape (28, 28, 1)
    layers.Conv2D(32, kernel_size=(3, 3), activation='relu',
                  input_shape=(28, 28, 1)),          # 32 filters, 3x3 kernel
    layers.MaxPooling2D(pool_size=(2, 2)),            # halves spatial dimensions

    layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),   # 64 filters
    layers.MaxPooling2D(pool_size=(2, 2)),

    layers.Flatten(),                                 # flatten to 1D vector
    layers.Dense(128, activation='relu'),             # fully connected layer
    layers.Dense(10,  activation='softmax')           # output: 10 classes
], name="CNN")

# Compile the CNN
cnn_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# 1. Print model summary — compare parameters with FFNN
print("\n" + "=" * 50)
print("CNN — Model Summary")
print("=" * 50)
cnn_model.summary()

cnn_params = cnn_model.count_params()
print(f"\nTotal Parameters (CNN)  : {cnn_params:,}")
print(f"Total Parameters (FFNN) : {ffnn_params:,}")
print(f"Difference              : {abs(cnn_params - ffnn_params):,}")

# 2. Train for 10 epochs with validation_split=0.2
print("\nTraining CNN...")
cnn_history = cnn_model.fit(
    X_train_cnn, y_train,
    epochs=10,
    validation_split=0.2,
    verbose=1
)

# 3. Evaluate CNN on test set
cnn_test_loss, cnn_test_acc = cnn_model.evaluate(X_test_cnn, y_test, verbose=0)
print(f"\nCNN Test Loss     : {cnn_test_loss:.4f}")
print(f"CNN Test Accuracy : {cnn_test_acc:.4f}  ({cnn_test_acc*100:.2f}%)")

# 4. Comparison Table — FFNN vs CNN
ffnn_train_acc = ffnn_history.history['accuracy'][-1]
ffnn_val_acc   = ffnn_history.history['val_accuracy'][-1]
cnn_train_acc  = cnn_history.history['accuracy'][-1]
cnn_val_acc    = cnn_history.history['val_accuracy'][-1]

comparison_df = pd.DataFrame({
    'Model'      : ['Feedforward NN', 'CNN'],
    'Parameters' : [f"{ffnn_params:,}", f"{cnn_params:,}"],
    'Train Acc'  : [f"{ffnn_train_acc:.4f}", f"{cnn_train_acc:.4f}"],
    'Val Acc'    : [f"{ffnn_val_acc:.4f}",   f"{cnn_val_acc:.4f}"],
    'Test Acc'   : [f"{ffnn_test_acc:.4f}",  f"{cnn_test_acc:.4f}"],
    'Epochs'     : [15, 10]
})

print("\n" + "=" * 60)
print("COMPARISON TABLE — Feedforward NN vs CNN")
print("=" * 60)
print(comparison_df.to_string(index=False))

# 5. Why CNNs outperform Dense networks on image data
print("\n" + "=" * 60)
print("Why CNNs Outperform Dense Networks on Image Data")
print("=" * 60)
print("""
A Dense (fully connected) layer treats every pixel as an independent
feature, destroying all spatial relationships between neighbouring pixels.
A Conv layer, by contrast, slides a small filter (kernel) across the image
and learns local patterns — edges, curves, corners — wherever they appear
(translation invariance). This spatial awareness is critical for images:
the digit '3' looks the same whether it's shifted left or right, but a
Dense layer would treat those as completely different inputs. CNNs also use
far fewer parameters (weight sharing across the kernel), making them more
efficient and less prone to overfitting on image data.
""")
```

A Conv layer looks at where pixels are relative to each other and learns local spatial patterns that repeat across the image — something a Dense layer is architecturally blind to, since it treats every pixel as if it exists in isolation.
