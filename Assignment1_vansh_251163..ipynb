```python
import seaborn as sns
df = sns.load_dataset('titanic')
print(df.head(5))
```

```python
df.info
```

```python
df.describe
```

```python
import pandas as pd
missing_value = pd.DataFrame({"Null count" : df.isnull().sum(), "Null Percentage": df.isnull().mean()*100})
print(missing_value)
```

```python
group = df.groupby("pclass")
print(group["age"].median())
```

```python
df['age'] = df['age'].fillna(group['age'].transform('median'))
```

```python
mode = df['embarked'].mode()[0]
print(mode)
df = df.fillna({'embarked': mode})
```

```python
df = df.drop(columns=['deck'])
```

Since deck column has maximum percentage of null values (i.e. 77%) in the dataset due to this our predictions becomes highly unreliable. There will be more noise in dataset rather than useful patterns

```python
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(15, 10))

#histogram
ax1 = axes[0]
ax1.hist(df['age'].dropna(), bins=20, color='steelblue', edgecolor='white')

mean_age = df['age'].mean()
median_age = df['age'].median()

ax1.axvline(mean_age, color='red', linestyle='--', linewidth=2, label=f'Mean Age: {mean_age:.1f}')
ax1.axvline(median_age, color='green', linestyle='-', linewidth=2, label=f'Median Age: {median_age:.1f}')

ax1.set_xlabel('Age')
ax1.set_ylabel('Frequency')
ax1.set_title('Distribution of Age')
ax1.legend()


#scatter plot
ax2 = axes[1]

not_survived = df[df['survived'] == 0]
survived = df[df['survived'] == 1]

ax2.scatter(not_survived['age'], not_survived['fare'], color='red', label='Not Survived (0)')
ax2.scatter(survived['age'], survived['fare'], color='green', label='Survived (1)')

ax2.set_ylim(0, 300)

ax2.set_xlabel('Age')
ax2.set_ylabel('Fare')
ax2.set_title('Age vs Fare')
ax2.legend(title="Status")

plt.tight_layout()
plt.show()
```

The y-axis on the Scatter Plot (`Age vs Fare`) is capped at a maximum value of 300. This removes extreme outward point which also enhance the visual readability and distribution mapping of the plot.

```python
train_df = df_adv.iloc[:160]
test_df = df_adv.iloc[160:]

X_train = train_df['TV'].values
y_train = train_df['Sales'].values

X_test = test_df['TV'].values
y_test = test_df['Sales'].values

x_bar = np.mean(X_train)
y_bar = np.mean(y_train)

numerator = np.sum((X_train - x_bar) * (y_train - y_bar))
denominator = np.sum((X_train - x_bar) ** 2)

m_scratch = numerator / denominator
b_scratch = y_bar - (m_scratch * x_bar)

print("Computed Slope (m):",m_scratch)
print("Computed Intercept (b):",b_scratch)

y_pred_scratch = (m_scratch * X_test) + b_scratch

mse_scratch = np.mean((y_test - y_pred_scratch) ** 2)
print("Manual Test MSE:",mse_scratch)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = df_adv[['TV']]
y = df_adv['Sales']

X_train_sk, X_test_sk, y_train_sk, y_test_sk = train_test_split(
    X, y, test_size=0.20, random_state=42
)

sk_model = LinearRegression()
sk_model.fit(X_train_sk, y_train_sk)

m_sklearn = sk_model.coef_[0]
b_sklearn = sk_model.intercept_
y_pred_sk = sk_model.predict(X_test_sk)
mse_sklearn = mean_squared_error(y_test_sk, y_pred_sk)

print("Sklearn Coefficient (m):",m_sklearn)
print("Sklearn Intercept (b):",b_sklearn)
print("Sklearn Test MSE:",mse_sklearn)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_train_pred = sk_model.predict(X_train_sk)
y_test_pred = sk_model.predict(X_test_sk)

mae_train = mean_absolute_error(y_train_sk, y_train_pred)
mse_train = mean_squared_error(y_train_sk, y_train_pred)
rmse_train = np.sqrt(mse_train)
r2_train = r2_score(y_train_sk, y_train_pred)

mae_test = mean_absolute_error(y_test_sk, y_test_pred)
mse_test = mean_squared_error(y_test_sk, y_test_pred)
rmse_test = np.sqrt(mse_test)
r2_test = r2_score(y_test_sk, y_test_pred)

print(f"{'Evaluation Metric':<18} | {'Train Set':<9} | {'Test Set':<9}")
print(f"{'MAE':<18} | {mae_train:<9.4f} | {mae_test:<9.4f}")
print(f"{'MSE':<18} | {mse_train:<9.4f} | {mse_test:<9.4f}")
print(f"{'RMSE':<18} | {rmse_train:<9.4f} | {rmse_test:<9.4f}")
print(f"{'R² Score':<18} | {r2_train:<9.4f} | {r2_test:<9.4f}")
```

```python
from sklearn.cluster import KMeans
df_mall = pd.read_csv("/content/Mall_Customers.csv")
print(df_mall.head())
print(df_mall.info())
print(df_mall.describe())
```

```python
plt.scatter(df_mall['Annual Income (k$)'], df_mall['Spending Score (1-100)'])
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Annual Income vs Spending Score')
```

I can spot 5 natural groups on the plot:
  1. Bottom-Left: Low income, low spending score.
  2. Top-Left: Low income, high spending score.
  3. Center: Average income, average spending score.
  4. Bottom-Right: High income, low spending score.
  5. Top-Right: High income, high spending score.

```python
from sklearn.cluster import KMeans

income_col = 'Annual Income (k$)'
spending_col = 'Spending Score (1-100)'

X_clustering = df_mall[[income_col, spending_col]].values

kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
df_mall['Cluster'] = kmeans.fit_predict(X_clustering)

centroids = kmeans.cluster_centers_
centroids_df = pd.DataFrame(centroids, columns=['Centroid Annual Income', 'Centroid Spending Score'])
centroids_df.index.name = 'Cluster ID'

print(centroids_df.to_string())

cluster_summary = df_mall.groupby('Cluster').agg(
    Customer_Count=(income_col, 'count'),
    Mean_Annual_Income=(income_col, 'mean'),
    Mean_Spending_Score=(spending_col, 'mean')
)

for cluster_id, row in cluster_summary.iterrows():
    print(f"Cluster",cluster_id)
    print(f"  • Count of Customers  :", (row['Customer_Count']))
    print(f"  • Mean Annual Income  : ${row['Mean_Annual_Income']:.2f}k")
    print(f"  • Mean Spending Score : {row['Mean_Spending_Score']:.2f}")
```
